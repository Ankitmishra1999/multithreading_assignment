```python
Q.1 what is multithreading in python? why is it used? Name the module used to handle threads in python

ans.
    Multithreading in Python refers to the concurrent execution of multiple threads within a single process.
    A thread is the smallest unit of execution within a process, 
    and multithreading allows you to run multiple threads concurrently to perform tasks concurrently or in parallel.
    Each thread has its own program counter, stack, and local variables,
    but they share the same memory space of the parent process, which can be both an advantage and a challenge.

Multithreading is used in Python for several purposes, including:

Concurrency: It allows you to perform multiple tasks simultaneously without blocking the main program's execution.
This can be beneficial for tasks that involve I/O operations, like reading and writing to files or making network requests, 
as threads can work on these tasks concurrently while the main program continues to run.

Parallelism: In Python, because of the Global Interpreter Lock (GIL), true multi-core parallelism is limited. 
However, for certain types of tasks (e.g., CPU-bound tasks), you can still achieve some degree of parallelism by using threads,
even if they don't fully utilize multiple CPU cores due to the GIL.


The module commonly used to handle threads in Python is the threading module.
    Here's a basic example of how to use it:
    
import threading

def worker_function():
    # This is the function that will be executed by the thread
    print("Worker thread is doing some work")

# Create a thread object
thread = threading.Thread(target=worker_function)

# Start the thread
thread.start()

# Wait for the thread to finish (optional)
thread.join()

print("Main program continues running")
    
```

```python
Q 2. why threading module used? rite the use of the following functions
1.activeCount()
2. currentThread()
3.enumerate()

ans.

The threading module in Python is used for managing and working with threads in a Python program.
It provides a high-level, object-oriented interface for creating and managing threads.
Here's a brief explanation of the functions you've mentioned from the threading module:
 
1.activeCount():

Use: This function is used to retrieve the current number of Thread objects that are alive
 (i.e., not yet terminated) in the program.
example

import threading

# Create and start some threads
thread1 = threading.Thread(target=some_function)
thread2 = threading.Thread(target=another_function)
thread1.start()
thread2.start()

# Get the count of active threads
active_count = threading.activeCount()
print(f"Number of active threads: {active_count}")



2.currentThread():

Use: This function returns a reference to the currently executing Thread object.
 You can use it to access information about the current thread.
example 
import threading

def my_function():
    current_thread = threading.currentThread()
    print(f"Current thread name: {current_thread.name}")

thread = threading.Thread(target=my_function)
thread.start()


3.enumerate():

Use: The enumerate() function returns a list of all currently alive Thread objects.
It's often used to inspect and manage the active threads in a program.
example
import threading

def some_task():
    pass

def another_task():
    pass

# Create and start some threads
thread1 = threading.Thread(target=some_task)
thread2 = threading.Thread(target=another_task)
thread1.start()
thread2.start()

# Get a list of all currently active threads
thread_list = threading.enumerate()

# Print information about each thread
for thread in thread_list:
    print(f"Thread Name: {thread.name}")
```

```python
Q.3 Explain the following functions
1.run()
2.start()
3.join()
4.isAlive()

ans. 

1.run():

Use: This method is not typically called directly by the programmer.
Instead, it is meant to be overridden in a custom thread class that inherits from threading.Thread. 
You provide the code you want the thread to execute within the run() method.

example
import threading

class MyThread(threading.Thread):
    def run(self):
        print("Thread is running")

my_thread = MyThread()
my_thread.start()  # This internally calls the run() method

2.start():

Use: The start() method is used to initiate the execution of a thread. It internally calls the run() method,
so you should never call run() directly if you want to start a new thread.

import threading

class MyThread(threading.Thread):
    def start(self):
        print("Thread is running")

my_thread = MyThread()
my_thread.start()  # This internally calls the run() method


3.join():

Use: The join() method is used to wait for a thread to complete its execution before allowing the main program to proceed. 
This is useful when you want to ensure that a thread has finished its task before continuing with the rest of the program.

example
import threading

def some_task():
    print("Thread is doing some work")

thread = threading.Thread(target=some_task)
thread.start()
thread.join()  # Wait for the thread to finish before proceeding
print("Main program continues running")

4.isAlive():

Use: The isAlive() method is used to check whether a thread is currently alive or still executing. 
It returns True if the thread is alive and False if it has finished executing.
example
import threading
import time

def some_task():
    time.sleep(2)

thread = threading.Thread(target=some_task)
thread.start()

while thread.isAlive():
    print("Thread is still running...")
    time.sleep(1)
print("Thread has finished")
```

```python
Q.4 write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes.

ans.

import threading

def print_squares():
    for i in range(1, 6):
        square = i ** 2
        print(f"Square of {i} is {square}")

def print_cubes():
    for i in range(1, 6):
        cube = i ** 3
        print(f"Cube of {i} is {cube}")

# Create two thread objects
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start both threads
thread1.start()
thread2.start()

# Wait for both threads to finish (optional)
thread1.join()
thread2.join()

print("Both threads have finished")
```

```python
Q.5 State advantages and disadvantages of multithreading.

ans.
Advantages of Multithreading:

1.Concurrency: Multithreading allows multiple tasks to execute concurrently, 
improving the overall program performance by utilizing available CPU resources more efficiently.

2.Responsiveness: Multithreaded programs can be more responsive to user input, 
as background tasks can run in separate threads, ensuring that the main thread remains responsive to user interactions.

3.Efficiency: For certain types of tasks, such as I/O-bound operations,
multithreading can lead to significant performance improvements by
allowing threads to work on other tasks while some are blocked waiting for I/O operations to complete.

4.Resource Sharing: Threads within the same process share the same memory space, 
which makes it easier to share data and resources between threads without the need for complex inter-process communication (IPC) mechanisms.

5.Modularity: Multithreading can lead to more modular and organized code by allowing you to separate different tasks or
components of a program into separate threads, improving code readability and maintainability.

Disadvantages of Multithreading:

C1.omplexity: Multithreaded programs can be more complex and challenging to develop, debug, and maintain compared to single-threaded programs. 
Managing thread synchronization and avoiding race conditions can be particularly tricky.

2.Race Conditions: Concurrent access to shared resources can lead to race conditions, 
where multiple threads access and modify data simultaneously, resulting in unpredictable and erroneous behavior.

3.Deadlocks: Poorly managed synchronization can lead to deadlocks,
where threads are stuck waiting for resources that are held by other threads,
causing the entire program to come to a halt.

4.Performance Overhead: The overhead of creating, managing, 
and switching between threads can sometimes outweigh the benefits of multithreading,
especially for simple or CPU-bound tasks.

5.Limited Parallelism in Python (GIL): In Python, the Global Interpreter Lock (GIL) restricts true multi-core parallelism when using threads.
This limitation can make it less effective for CPU-bound tasks that would benefit from utilizing multiple CPU cores.

6.Debugging Challenges: Debugging multithreaded programs can be more challenging due to the 
non-deterministic nature of thread execution and potential timing-related issues.

7.Portability: Multithreading behavior can vary between different operating systems and platforms,
making it harder to write cross-platform code that behaves consistently.
```

```python
Q6. Explain deadlocks and race conditions.

ans.

Deadlock:
A deadlock is a situation in which two or more threads or processes are unable to proceed because they are each waiting for the other(s) to release a resource or complete a task. 
In other words, it's a state where two or more entities are stuck in a circular waiting pattern, preventing them from making any progress.

There are four necessary conditions for a deadlock to occur, often referred to as the "Deadlock Conditions":

1.Mutual Exclusion: At least one resource must be non-sharable, meaning that only one thread or process can access it at a time.

2.Hold and Wait: A process/thread must be holding at least one resource while waiting to acquire another resource.

3.No Preemption: Resources cannot be forcibly taken away from a process/thread. They can only be released voluntarily.

4.Circular Wait: There must exist a circular chain of two or more processes/threads, each waiting for a resource held by the next one in the chain

Race Condition:
A race condition occurs when the behavior of a program depends on the relative timing of events, such as when multiple threads or processes access shared resources concurrently without proper synchronization.
In a race condition, the outcome of the program is unpredictable because it depends on the order of execution of the threads or processes.

Race conditions typically happen when:
    
1.Multiple threads or processes access shared resources (e.g., variables, data structures) without proper synchronization mechanisms like locks or semaphores.

2.At least one of the threads or processes performs a write operation (e.g., modifying a shared variable), and there is no guarantee of atomicity or exclusivity in this operation.

3.The order of execution of the threads or processes is not deterministic, so different runs of the program may produce different results.    
```
